
# Electric Wave Propagation

## The Maxwell's Equarions
The Maxwell Equations can be decomposed into a set of second order independent equations in the Electirc field E and magnetic flux H, here we are interested into simulationg only the electric wave, therefore:

$$
\begin{cases}
\partial^2 _{tt} E  = \nabla \nabla E\\
\nabla \cdot D =0\\
D = \varepsilon E \\
\end{cases}
$$


If we set $d = \partial_t D$ and $e = E$ (for symmetry's sake) , and supposing that the properties of the material do not vary in time then we can decompose the previous equation into

$$
\begin{cases}
\varepsilon e_t  = d\\
d_t = - \text{curl curl }e \\
\text{ div } d = 0
\end{cases}
$$

## The Weak Form of Maxwell's Equarions

If we set our problem in the correct space $H(curl)\times H(div) \times L^2/\mathbb(R)$ then we can say that

$$
\begin{cases}
(\varepsilon e_t,u)  = (d,u)\\
(d_t,v) = (- \text{curl curl }e,v) \\
(\text{ div } d,q) = 0
\end{cases}
$$

$\forall (u,v,q) \in H(curl)\times H(div)\times  L^2/\mathbb(R)$
An important observation is that the above are generally eqautions in a dual space with respect to the space just mentioned, this remark will be usefull later since instead of solving the equation in the dual space $V^*$ we notice that it is easier to solve it in the predual space $V$ and then to apply the Hilbert functional to identify somehow the result from the predual to the dual.

In particular this form is not completely exact since in general the interplay between the various spaces is not defined in a symple $L^2$-like integration. Therefore it is more apropriate to define the problem in the following way:

### Correct Maxwell's Problem
------------------------------------
$$
\begin{cases}
(\varepsilon e_t,u)  = (\tilde{e}(t),u)\\
(d_t,v) = (\tilde{d}(t),v) \\
\end{cases}
$$

Where $\tilde{e}(t)\in [H(curl)]^*$ solves at every instant in time the following equation :

$$
\min_{[H(curl)]^*} \| \tilde{e}(t) - d\|^2
$$

and $\tilde{d}(t)\in [H(div)]^*$ solves at every instant in time the following equation :

$$
\min_{[H(div)]^*\ni \text{div}(\tilde{d}(t)) =0} \| \tilde{d}(t) - \text{curl curl} e\|^2
$$


$\forall (u,v,q) \in H(curl)\times H(div)\times L^2/\mathbb(R)$


----------------------------------
The remark made before is now taken into account, instead of finding $\tilde{e}$ we find $\hat{e}$ and we pass back using the Hilbert functional:
$$
\tilde{e} = \left.\mathcal{J}\right. _{H(curl)}^{H(curl)^*}\hat{e}
$$

Now we will make an important usage of the following:
1. It is possible to prove that the norm does not have to hold for the respective graph norms of the corrispective spaces but it needs only to hold for $L^2$, then for example $$ \min_{\hat{e}\in H(curl)}\| \hat{e}-f\|_{H(curl)} = \min_{\hat{e}\in H(curl)}\| \hat{e}-f\|_{L2} $$
2. I suppose that for all hilbert space $V,W$ it holds true that : $$ \mathcal{J}^{-1} \left. P\right. _{V^*}^{W^*}g =\left. P\right. _{V}^{W}\mathcal{J}^{-1}g$$

### Better Maxwell's problem

----------------------------------
Find 
$$
\begin{cases}
(\varepsilon e_t,u)  = (\tilde{e}(t),u)\\
(d_t,v) = (\tilde{d}(t),v) \\
\end{cases}
$$

Where the $\tilde{\cdot}$ functions are identified to the $\hat{\cdot}$ via the previous defined hilbert functional and
Where $\hat{e}(t)\in H(curl)$ solves at every instant in time the following equation :

$$
(\hat{e}(t),u)=( \mathcal{J}^{-1}d, u)
$$

and $\hat{d}(t)\in H(div)$ solves at every instant in time the following equation :

$$
\begin{cases}
(\hat{d}(t),v) &+ (div(v), p) &= (\mathcal{J}^{-1}\text{curl curl} e,v) \\
(div(v), p)& &=0
\end{cases}
$$


$\forall (u,v,q) \in H(curl)\times H(div)\times L^2/\mathbb(R)$

------------------------------
Therefore 

Since $(e,d)\in H(curl)\times H(div)$ we need to perform an $L_2$-projection
we need to solve at each time step $n$ the following projection: 
$$
\begin{cases}
\hat{e}^n = e^{n}\\
\hat{d}^n = \text{curl curl }e^n \\
\text{div }\hat{d}^n = 0\\
\end{cases}
$$
and:
$$
\begin{cases}
e^{n+1} & = e^{n} -\tau \hat{e}^n \\
d^{n+1}  &= d^{n} -\tau \hat{d}^n \\
\end{cases}
$$

### Numerical implementation:
When we apply the Galerkin discretization we obtain something of the following form:

$$
\begin{cases}
M_{e} \,\hat{\underline e}^n & = M_{ed}\, M_d^{-1} \, \underline d^{n}\\
M_{d} \hat{\underline d}^n &=  M_{de}  M_e^{-1}  \, K \, \underline e^n \\
\text{div }\hat{\underline d}^n& = 0\\
\end{cases}
$$
And to step foreward transforms form:
$$
\begin{align*}
\underline e^{n+1} & = \underline e^{n} -\tau \hat{\underline e}^n \\
\underline d^{n+1}  &= \underline d^{n} -\tau \hat{\underline d}^n \\
\end{align*}
$$

to 

$$
\begin{align*}
\underline e^{n+1} & = \underline e^{n} -\tau M_{e}^{-1} M_{ed}\, M_d^{-1} \,  \underline d^{n} \\
\underline d^{n+1}  &= \underline d^{n} -\tau  M_{d}^{-1}  M_{de}  M_e^{-1}  \, K \, \underline e^n \\
\underline d^{n+1}  &= P \, \underline d^n \\
\end{align*}
$$
Where the last one is the divergence free formjection of $d$.


### We need tobuild the matrix:
$$
\begin{pmatrix}
0 & M_{e}^{-1} M_{ed}\, M_d^{-1}  \\
 M_{d}^{-1}  M_{de}  M_e^{-1} K & 0 
\end{pmatrix}
$$

```python
def MaxwellPropagation(mesh , order = 1 , initail_E_field = None , bonus_intorder = 10 ,draw_e_field = True , draw_d_field = False, draw_divd_field = False , \
                    final_time  = 0.1, step_time = 0.002 , MoreInfo = False,video = False):
    
    # define the 3 spaces we need
    if MoreInfo:
        start = time.time()
        print("create the spaces :")

    fes =  HCurl(mesh, order=order )*HDiv(mesh, order=order   , dirichlet= ".*")
    fes_projection =  HDiv(mesh, order=order   , dirichlet= ".*")*L2(mesh,  order=order-1)

    if MoreInfo:
        end = time.time()
        print(str(end-start)+"\n")

    #electric field
    if MoreInfo:
        start = time.time()
        print("interpolate the initaial electric field :")

    elfield = GridFunction(fes) 
    e, d = elfield.components
    e.Set ( initail_E_field, bonus_intorder=10, dual=True)

    if MoreInfo:
        end = time.time()
        print(str(end-start)+"\n")

    # projectional part
    Pelfield = GridFunction(fes_projection) 
    Pd, Pq = Pelfield.components

    # trial and test function for H(curl) X H(div)
    uc, ud , = fes.TrialFunction()
    vc, vd , = fes.TestFunction()

    # create the inverse of the submatrix M_e 
    if MoreInfo:
        start = time.time()
        print("create the inverse of the submatrix M_e  :")

    inv_e = InverseBlockDiagonal( HCurl(mesh, order=order ))
    edofs = fes.Range(0)
    emb_e = Embedding(fes.ndof, edofs)

    if MoreInfo:
        end = time.time()
        print(str(end-start)+"\n")

    #create the inverse of M_d 
    if MoreInfo:
        start = time.time()
        print("create the inverse of the submatrix M_d  :")
        
    inv_d = InverseBlockDiagonal( HDiv(mesh, order=order ) )

    ddofs = fes.Range(1)
    emb_d = Embedding(fes.ndof, ddofs)

    if MoreInfo:
        end = time.time()
        print(str(end-start)+"\n")

    # create the passage matrix M_de : H(curl) -> H(div)
    if MoreInfo:
        start = time.time()
        print("create the passage matrix M_de : H(curl) -> H(div)  :")
    M_de = MixedBlockMatrix(HCurl(mesh, order=order ),HDiv(mesh, order=order   , dirichlet= ".*"))
    #c = HCurl(mesh, order=order ).TrialFunction()
    #v = HDiv(mesh, order=order   , dirichlet= ".*").TestFunction()
    #M_de = BilinearForm(trialspace=HCurl(mesh, order=order ), testspace=HDiv(mesh, order=order   , dirichlet= ".*") , geom_free = True)
    #M_de +=  InnerProduct(c, v)*dx
    #M_de.Assemble()
    input("so far at M_de")
    if MoreInfo:
        end = time.time()
        print(str(end-start)+"\n")


    # create the passage matrix M_ed : H(curl) -> H(div)
    input("so far at M_ed")
    if MoreInfo:
        start = time.time()
        print("create the passage matrix M_ed : H(curl) -> H(div)  :")
    M_ed = MixedBlockMatrix(HDiv(mesh, order=order   , dirichlet= ".*") , HCurl(mesh, order=order ))
    #c = HCurl(mesh, order=order ).TestFunction()
    #v = HDiv(mesh, order=order   , dirichlet= ".*").TrialFunction()
    #M_ed = BilinearForm(testspace=HCurl(mesh, order=order ), trialspace=HDiv(mesh, order=order   , dirichlet= ".*"), geom_free = True)
    #M_ed +=  InnerProduct(c, v)*dx
    #M_ed.Assemble()
    input("so far at M_ed")

    if MoreInfo:
        end = time.time()
        print(str(end-start)+"\n")

    # creating the matrix for the operator curl-curl:
    if MoreInfo:
        start = time.time()
        print("create the passage matrix for the operator curl-curl  :")

    u = HCurl(mesh, order=order ).TestFunction()
    v = HCurl(mesh, order=order ).TrialFunction()
    K = BilinearForm(HCurl(mesh, order=order ))
    K += -InnerProduct( curl(u), curl(v) )*dx
    K.Assemble()

    if MoreInfo:
        end = time.time()
        print(str(end-start)+"\n")


    # compose the 2 final matrices
    if MoreInfo:
        start = time.time()
        print("create the passage final matrices Mup Mdown  :")

    Mup = emb_e @ inv_e @ M_ed.mat @ emb_d.T
    Mdown = emb_d @ inv_d @ M_de.mat @ inv_e  @ K.mat @ emb_e.T 
    #M =  Mdown
    #   
    if MoreInfo:
        end = time.time()
        print(str(end-start)+"\n")

    # Create the projection operator for the divergence free part
    if MoreInfo:
        start = time.time()
        print("create the projection operator for the divergence free part  :")
    u, p  = fes_projection.TrialFunction()
    v, q  = fes_projection.TestFunction()

    P = BilinearForm(fes_projection)
    P +=  v*u * dx
    P +=  div( v)*p * dx
    P +=  div( u)*q * dx
    P += -1e-10*p*q * dx

    F = BilinearForm(fes_projection)
    F +=  v*u * dx

    P.Assemble()
    F.Assemble()

    invP = P.mat.Inverse(inverse="sparsecholesky")

    if MoreInfo:
        end = time.time()
        print(str(end-start)+"\n")

    if draw_e_field :
        scene_e = Draw(e, mesh,clipping=(0,0,1))
    if draw_divd_field :
        scene_divd = Draw(div(d), mesh,clipping=(0,0,1) ,draw_surf=False)
    if draw_d_field :
        scene_d = Draw(d, mesh,clipping=(0,0,1))
    
    if MoreInfo:
        start = time.time()
    t =   final_time
    tau = step_time
    h = 0
  
    
    TM =TaskManager()
        
    with TM:
      while h < t :
        h+=tau
        input("so far at h "+str(h))
        #we use only one matrix of the form described in the above markdowm
        e.vec.data += tau * (inv_e @ M_ed.mat)*d.vec
        d.vec.data += tau * (inv_d @ M_de.mat @ inv_e  @ K.mat)*e.vec

        Pd.vec.data = d.vec
        Pelfield.vec.data =  invP*(F.mat*Pelfield.vec)
        d.vec.data = Pd.vec

        if draw_e_field :
            scene_e.Redraw()
        if draw_d_field :
            scene_divd.Redraw()
        if draw_d_field :
            scene_d.Redraw()
    if MoreInfo :
        end = time.time()
        print("Time of execution for "+str(int(final_time/step_time)) + "steps is "+str(end-start))
    return e,d
```

```python
final_time = 0.3
step_time = 0.002
MoreInfo = True
video = False

Maxh = 0.13
order = 3
mesh = Mesh(unit_cube.GenerateMesh(maxh=Maxh))

# electric permettivity and magnetic permeability
eps = CF((1), dims=(1,1))
mu =  CF((1), dims=(1,1))

# define initial conditions for the wave
peak = exp(-35*( (x-0.5)**2 + (y-0.5)**2 +(z-0.5)**2))
initail_E_field = CF ( (0, 0, peak ), dims=(3,1))


e,d = MaxwellPropagation(mesh , order = order ,initail_E_field = initail_E_field , final_time  = final_time, step_time = step_time,MoreInfo = MoreInfo, video =video)
```
